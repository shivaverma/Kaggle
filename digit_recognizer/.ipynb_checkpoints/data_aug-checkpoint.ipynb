```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras import optimizers
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

print(train_data.shape, test_data.shape)
```

```python
labels = train_data['label']
train_data.drop('label', axis=1, inplace=True)
```

```python
def preprocess(data):
    
    data = data/255
    data -= 0.5
    data = data.values.reshape(data.shape[0],28,28,1)
    return data
```

```python
train_data = preprocess(train_data)
test_data = preprocess(test_data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_data, labels, test_size=0.2, random_state=12)

print(x_train.shape)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))                        # 13*13
model.add(tf.keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'))     # 12*12

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(10, activation='softmax'))

adam = tf.keras.optimizers.Adam()
model.compile(loss='sparse_categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
for i in range(100):
    
    plt.imshow(x_train[i].reshape((28,28)))
    plt.show()
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rotation_range=30, width_shift_range=.1, height_shift_range=.1,
                             shear_range=10, zoom_range=[.9, .9])
```

```python
datagen.fit(x_train)
```

```python
epoch = 0
batch = 128
max_epoch = 2

model.optimizer.lr = .0001

for x,y in datagen.flow(x_train, y_train, batch_size=batch):
    
    epoch += 1
    model.fit(x_train, y_train, batch_size=batch)
    if epoch == max_epoch:
        break
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print("\nval loss:", val_loss, "\nval_acc:", val_acc)
```
